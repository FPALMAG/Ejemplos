## Potencial de Upside Sectorial en empresas del S&P500

```python
print('El script empieza seleccionando la lista de empresas del S&P500','\n','obteniendo su simbolo o Ticker')
```

```python
import yfinance as yf
import pandas as pd
import time
```

```python
tickers=['AAPL','MSFT','VZ']
```

```python
datos={'Ticker':[],'Beta':[],'Precio':[]}
```

```python
for i in tickers:
    df=yf.Ticker(i).info
    datos['Ticker'].append(i)
    datos['Beta'].append(df['beta'])
    datos['Precio'].append(df['previousClose'])
    print(datos)
```

```python
datos
```

# Para mayor cantidad de datos

```python
lista_sp500=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
lista_sp500
```

```python
sp500_ticker_list=list(lista_sp500.Symbol)
sp500_ticker_list
```

```python
ratios={'Ticker':[],'Sector':[],'Precio':[],'Precio Objetivo':[]}
```

```python
for i in sp500_ticker_list:
    df_sp500=yf.Ticker(i).info
    if 'sector' in df_sp500:
        ratios['Sector'].append(df_sp500['sector'])
    else:
        ratios['Sector'].append(None)
    ratios['Ticker'].append(i)
    if 'currentPrice' in df_sp500:
        ratios['Precio'].append(df_sp500['currentPrice'])
    else:
        ratios['Precio'].append(None)
    if 'targetMeanPrice' in df_sp500:
        ratios['Precio Objetivo'].append(df_sp500['targetMeanPrice'])
    else:
        ratios['Precio Objetivo'].append(None)
    time.sleep(2)
    #print(ratios)
```

```python
ratios
```

```python
ratios_df=pd.DataFrame(ratios)
ratios_df
```

```python
ratios['Upside']=(ratios_df['Precio Objetivo']/ratios_df['Precio']-1)*100
```

```python
ratios
```

```python
df_ratios=pd.DataFrame(ratios)
df_ratios
```
